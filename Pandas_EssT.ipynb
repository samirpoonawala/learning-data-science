### Using value_counts()

* `Series.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)`

```
oo.Gender.value_counts()
```

### Basic plotting

```
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plot types
* plot()
  * plot(kind='line') - default
    * best for tracking changes over a period of time
  * plot(kind='bar')
    * best for comparing between different groups
    * best when changes are large
  * plot(kind='barh')
    * same as above but with axes flipped
  * plot(kind='pie')
    * best for comparing parts of a whole

```
# What were the different sports in the first olympics?

fo = oo[oo.Edition == 1896]
fo.Sport.value_counts().plot(kind='bar');
```

### Colors

* plot(kind='bar', color='red')
* Named Matplotlib colors [link](http://matplotlib.org/color/named_colors.html)

```
fo.Sport.value_counts().plot(kind='bar', color='blue');
```

### Figsize
* plot(figsize=(width, height))
  * plot(kind='bar', color='blue', figsize=(5,5))

```
fo.Sport.value_counts().plot(kind='bar', color='blue', figsize=(12,3))
```

### Colormaps
* Classes of colormaps:
  * sequential
    * use for information that has ordering
    * change in lightness, often over a single hue
  * diverging
    * use when information plotted deviates around a middle value
  * qualitative
    * used to represent information that does not have any ordering or relationship

```
fo.Sport.value_counts().plot(kind='pie', colormap='Pastel1')
```

## (7) Groupby

### Groupby
* How groupby() works:
  * split a dataframe into groups based on some criteria
  * apply a function to each group independently
  * combine the results into a dataframe
* Returns a groupby object

```
oo.groupby('Edition')
```

### Iterate through a group
* for key, group in DataFrame.groupby():
  * print(key)
  * print(group)

```
for group_key, group_value in oo.groupby("Edition"):
  print(group_key)
  print(group_value)
```

```
type(group_value)
```

### Groupby computations
* GroupBy.size()
* GroupBy.count()
* GroupBy.first(), GroupBy.last()
* GroupBy.head(), GroupBy.tail()
* GroupBy.mean()
* GroupBy.max(), GroupBy.min()

On the job:
* agg() - multiple statistics in one calculation per group

```
oo.groupby('Edition').size()
```

```
oo.groupby(['Edition', 'NOC', 'Medal']).agg(['min','max','count'])
```

```
# using dict like structure gives a clearer output here
# oo.groupby(['Edition','NOC','Medal']).agg({'Edition': ['min','max','count']})

oo.groupby(['Edition','NOC','Medal']).agg({'Edition': ['count']})
```

```
oo.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({"Edition":['min','max','count']})
```

## (8) Reshaping

### Reshaping
* Stacking
* Unstacking

```
mw = oo[(oo.Edition == 2008) & ((oo.Event == "100m") | (oo.Event == '200m'))]
mw.head()
```

```
g = mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```
df = g.unstack(['Discipline','Event'])
df
```

### Using stack()

* Pivot a new level of the column labels, returning a DataFrame or Series, with a new innermost level of row labels
* Returns a newn DataFrame (or Series) that is taller

```
# stacking gives you a taller dataframe
df.stack()
```

```
df.stack("Event")
```

### Using unstack()

* Pivot a level of the index labels, returning a DataFrame having a new level of column labels
* Returns a wider dataframe

```
df.unstack()
```

```
df.unstack("Gender")
```

### Challenge
* Plot the number of gold medals won by the US male and female athletes throughout the history of the olympics
* Plot the five athletes who have won the most gold medals over the history of the Olympics. When there's a tie, consider the number of silver medals, and then bronze medals

```
# plot number of gold medals won by US male and female athletes
gu = oo[(oo.NOC == "USA") & (oo.Medal == 'Gold')]

gu.groupby(['Edition','Gender']).size().unstack("Gender", fill_value=0).plot()
```

```
# plot the 5 athletes with most gold medals over history of olympics

g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']].head(5).plot(kind='bar')
```

## (10) Final challenge

* In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline

```
gy = oo[oo.NOC == "USA"]
gy = gy.groupby(['Edition','Athlete','Medal']).size().unstack('Medal', fill_value=0)

gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']

gy.reset_index(inplace=True)

tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')]
tu

top = pd.DataFrame()
for i in tu:
  top = top.append(i)

top
```
